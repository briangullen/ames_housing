## Question 8: <a class="anchor" id="Question8"></a>
### what are the homes in Gilbert with Year Sold being 2009 and 2010?

```python
## Neighborhood: Gilbert, Years: 2009 & 2010

results_df = housing_df[( (housing_df["Neighborhood"]=="Gilbert") &  
    (housing_df["Yr Sold"]==2009) ) |
    (housing_df["Yr Sold"]==2010)]
display(results_df)
```

## Question 9: <a class="anchor" id="Question9"></a>
### What is the count of home sold in Gilbert with Yr Sold being 2010

```python
## Neighborhood: Gilbert, Years: 2010
results_df = housing_df.loc[
    (housing_df["Neighborhood"]=="Gilbert") &
    (housing_df["Yr Sold"]==2010)]

display(results_df['Yr Sold'].value_counts())
```
